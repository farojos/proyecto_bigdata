```python
with open('../feliperojos/Desktop/Apple-Twitter-process.txt','r') as f:
    output = f.read()
```

```python
out = output.split('\n')
```

```python
import re
new_f = []
problems = []
for i in range(len(out)):
    if(re.match('\d,',out[i])):
        pass #print('yay',i)
        
    else:
        problems.append(i)
    
```

```python
better_problems = []
inv = [problems[0]]
for i in range(len(problems)):
    if(i+2<len(problems)):
        if(problems[i+1]-problems[i]==1):
            inv.append(problems[i+1])
        else:
            better_problems.append(inv)
            inv=[problems[i+1]]
        
    
```

```python
for i in better_problems:
    fe = i[0]
    start = fe-1
    for j in i:
        out[start]+= out[j]
```

```python
for i in range(len(out)):
    if(i in problems):
        pass #print('yay',i)
        
    else:
        new_f.append(out[i])
```

```python
final_f = []
for i in new_f:
    sent = int(i[:1])
    if(sent == 5 or sent == 4 or sent == 3):
        sent=1
    else:
        sent=0
    comment = i[2:]
    comment = re.sub('[^A-z0-9 ]', '', comment).lower()
    final_f.append(str(sent)+', '+comment)
```

```python
final_f 
```

```python
file = open('preprocesado.txt', 'w')
for item in final_f:
    file.write("%s\n" % item)
file.close()
```
